```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv")
print(type(df.day[0]))
```

```python
new_df = df.set_index('day')
new_df
```

```python
df.set_index("day", inplace = True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed': 0,
    'event': 'No Event'
})
new_df
```

```python
new_df = df.fillna(method="ffill") #ffill limit 
new_df
```

```python
new_df = df.fillna(method="ffill", limit = 3) #ffill limit 
new_df
```

```python
new_df = df.fillna(method="ffill", axis = 1) #ffill limit 
new_df
```

```python
new_df = df.fillna(method = 'bfill')
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
df
```

```python
new_df = df.dropna(how = 'all')
new_df
```

```python
new_df = df.dropna(thresh = 1)
new_df
```

```python
new_df = df.dropna(thresh = 2)
new_df
```

```python
new_df = df.dropna(thresh = 0.3*len(df.columns)) #non-empty values with percentage
new_df
```

```python
new_df = df.dropna(thresh = 7, axis = 1) #7 non-nan values
new_df
```
